```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Logloss (функция потерь)

```python
values = [-25, -10,-3, -2, -1, 0, 1, 2, 3, 10, 25]
y = []
for i in values:
    result = 1 / (1 + math.e**(-i))
    print(i, '<>', round(result, 5))
    y.append(result)

print()
```

```python
sns.scatterplot(x=values, y=y)
```

```python
import math

values = [-1, -0.5, -0.1, 0, 0.1, 0.5, 1]
y = []
for i in values:
    result = 1 / (1 + math.e**(-i))
    print(i, '<>', round(result, 5))
    y.append(result)

print()
sns.scatterplot(x=values, y=y)
```

## Энтропийный критерий

```python
values = np.arange(0.1, 1, 0.1)
y = []
for i in values:
    result = -(i * np.log2(i)) - (1-i)*(np.log2(1-i))
    print(i, '<>', round(result, 5))
    y.append(result)
```

```python
ax = sns.scatterplot(x=values, y=y)
ax.set(xlabel='∑ pi*log2(pi)', ylabel='Энтропия');
```

```python
import numpy as np
np.random.seed(7)
train_data = np.random.normal(size=(100, 2))
train_labels = np.zeros(100)

# добавляем второй класс
train_data = np.r_[train_data, np.random.normal(size=(100, 2), loc=2)]
train_labels = np.r_[train_labels, np.ones(100)]
```

```python
plt.rcParams['figure.figsize'] = (10,8)
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels, s=100, cmap='autumn', edgecolors='black', linewidth=1.5);
plt.plot(range(-2,5), range(4,-3,-1));
```

```python
def get_grid(data):
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

get_grid(np.arange(10).reshape(2,5))
```
