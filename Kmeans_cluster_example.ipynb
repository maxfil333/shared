```python
# импортируем необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris(as_frame=True).data
df
```

```python
# загружаем данные для кластеризации
data = df

# создаем объект модели KMeans с количеством кластеров n_clusters = 3
kmeans = KMeans(n_clusters=3)

# обучаем модель на данных
kmeans.fit(data)

# получаем метки кластеров для каждого объекта в наборе данных
labels = kmeans.predict(data)

# получаем координаты центроидов каждого кластера
centroids = kmeans.cluster_centers_
```

```python
# выводим результаты кластеризации на графике
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels, s=50, cmap='cool')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

```python
# выводим результаты кластеризации на графике
plt.scatter(data.iloc[:, 1], data.iloc[:, 2], c=labels, s=50, cmap='cool')
plt.scatter(centroids[:, 1], centroids[:, 2], c='black', s=200, alpha=0.5)
plt.show()
```
