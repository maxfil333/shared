```python
fig = plt.figure(figsize=(7,4))
ax_3d = fig.add_subplot(projection='3d')

x = np.arange(-2*np.pi, 2*np.pi, 0.2)
y = np.arange(-2*np.pi, 2*np.pi, 0.2)
xgrid, ygrid = np.meshgrid(x, y)

zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

# ax_3d.plot_wireframe(xgrid, ygrid, zgrid)
ax_3d.plot_surface(xgrid, ygrid, zgrid, cmap='plasma')
plt.show()
```

```python
# plt.figure() 
#    Figure ( )

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)  #    (1x1 ,  )
ax.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# fig.add_subplot()
#   (Axes)      .
# : fig.add_subplot(nrows, ncols, index) (, 111 = 1x1 ,  1)

fig = plt.figure()
ax1 = fig.add_subplot(211)  # 2 , 1 , 1- 
ax2 = fig.add_subplot(212)  # 2 , 1 , 2- 
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([1, 2, 3], [6, 5, 4])
plt.show()
```

```python
# plt.subplots()
# :        Axes   .
# : fig, axes = plt.subplots(nrows, ncols, **kwargs).

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 4))
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([1, 2, 3], [6, 5, 4])
plt.show()
```

```python
# plt.subplot2grid()  
# :        ,     ( GridSpec).
# : plt.subplot2grid(shape, loc, rowspan=1, colspan=1).

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)  #    
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)  #  2    
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)  #  2    
ax1.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# plt.GridSpec 
# :        .
# :   fig.add_subplot(gs[i:j, k:l])       .

import matplotlib.gridspec as gridspec
fig = plt.figure()
gs = gridspec.GridSpec(3, 3)
ax1 = fig.add_subplot(gs[0, :])  #  ,  
ax2 = fig.add_subplot(gs[1, :2])  #  ,   
ax3 = fig.add_subplot(gs[1:, 2])  #    ,  
ax1.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# plt.contour(), plt.contourf()  
# :    (   ).

import numpy as np
X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = np.sin(X) * np.cos(Y)
plt.contourf(X, Y, Z)
plt.show()
```

##  Pandas (Series.plot(), DataFrame.plot())

```python
# Series.plot()
# :        Series.
#  : kind='line', 'bar', 'hist', 'box', 'pie', 'scatter',  .
# :      X,   Matplotlib  **kwargs.

import pandas as pd
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.plot(kind='line')  #  .plot.line()
plt.show()
```

```python
# DataFrame.plot()  
# :      DataFrame.
# :       .

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df.plot(x='A', y='B', kind='scatter')
plt.show()
```
