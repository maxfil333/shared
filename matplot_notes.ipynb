# Новое

Основные подходы к созданию графиков в MatplotlibMatplotlib имеет два основных интерфейса для создания графиков:

- Pyplot (процедурный интерфейс): Использует модуль plt для быстрого создания графиков. Удобен для простых задач, но менее гибок для сложных компоновок.
- Объектно-ориентированный интерфейс: Работа с объектами Figure и Axes напрямую. Более гибкий, подходит для сложных графиков с несколькими подграфиками.



## Figure, Axes (ООП подход)

Создание фигуры и осей (Axes)
- Эти методы создают объект Figure (холст) и объект Axes (область для рисования графиков).plt.figure()  
- Назначение: Создает новый объект Figure (пустой холст).
- Использование: Обычно используется в ОО-подходе, когда нужно явно управлять фигурой.

```python
# plt.figure() 
# Создает новый объект Figure (пустой холст)

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)  # Добавляет одну область (1x1 сетка, первая позиция)
ax.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# fig.add_subplot()
# Добавляет подграфик (Axes) к существующей фигуре в виде сетки.
# Синтаксис: fig.add_subplot(nrows, ncols, index) (например, 111 = 1x1 сетка, позиция 1)

fig = plt.figure()
ax1 = fig.add_subplot(211)  # 2 ряда, 1 столбец, 1-я позиция
ax2 = fig.add_subplot(212)  # 2 ряда, 1 столбец, 2-я позиция
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([1, 2, 3], [6, 5, 4])
plt.show()
```

```python
# plt.subplots()
# Назначение: Создает фигуру и сразу возвращает массив объектов Axes для заданной сетки.
# Синтаксис: fig, axes = plt.subplots(nrows, ncols, **kwargs).

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 4))
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([1, 2, 3], [6, 5, 4])
plt.show()
```

```python
# plt.subplot2grid()  
# Назначение: Создает подграфик в сетке с более гибким расположением, включая возможность объединения ячеек (аналог GridSpec).
# Синтаксис: plt.subplot2grid(shape, loc, rowspan=1, colspan=1).

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)  # Занимает весь первый ряд
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)  # Занимает 2 ячейки во втором ряду
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)  # Занимает 2 ряда в третьем столбце
ax1.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# plt.GridSpec 
# Назначение: Создает объект сетки для точного управления расположением подграфиков.
# Использование: Используется с fig.add_subplot(gs[i:j, k:l]) для размещения подграфиков в определенных ячейках сетки.

import matplotlib.gridspec as gridspec
fig = plt.figure()
gs = gridspec.GridSpec(3, 3)
ax1 = fig.add_subplot(gs[0, :])  # Первый ряд, все столбцы
ax2 = fig.add_subplot(gs[1, :2])  # Второй ряд, первые два столбца
ax3 = fig.add_subplot(gs[1:, 2])  # Второй и третий ряды, третий столбец
ax1.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

## Методы для быстрого построения графиков (Pyplot)

Эти методы автоматически создают фигуру и оси, если они не заданы, и предназначены для простых графиков.

```python
# plt.plot() 
# Назначение: Рисует линию или точки на графике.
# Особенности: Автоматически создает Figure и Axes, если их нет.

plt.plot([1, 2, 3], [4, 5, 6], label='Line')
plt.legend()
plt.show()
```

```python
# plt.scatter()  
# Назначение: Создает точечный график.

plt.scatter([1, 2, 3], [4, 5, 6], color='red')
plt.show()
```

```python
# plt.bar(), plt.barh()
# Назначение: Создает столбчатые графики (вертикальные или горизонтальные).

plt.bar(['A', 'B', 'C'], [4, 3, 5])
plt.show()
```

```python
# plt.hist()  
# Создает гистограмму.

plt.hist([1, 2, 2, 3, 3, 3], bins=3)
plt.show()
```

```python
# plt.contour(), plt.contourf()  
# Назначение: Создает контурные графики (линии или заполненные области).

import numpy as np
X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = np.sin(X) * np.cos(Y)
plt.contourf(X, Y, Z)
plt.show()
```

```python
# plt.imshow()
# Назначение: Отображает двумерные данные как изображение (например, тепловые карты).

plt.imshow([[1, 2], [3, 4]], cmap='viridis')
# plt.colorbar()
# plt.show()
```

```python
# plt.boxplot() 
# Назначение: Создает коробчатую диаграмму (box-and-whisker).

plt.boxplot([[1, 2, 3], [4, 5, 6]])
plt.show()
```

```python
# plt.pie()
# Назначение: Создает круговую диаграмму.

plt.pie([30, 40, 30], labels=['A', 'B', 'C'])
plt.show()
```

## Методы Pandas (Series.plot(), DataFrame.plot())



Pandas использует Matplotlib как бэкенд для визуализации данных, предоставляя удобный интерфейс для построения графиков.



```python
# Series.plot()
# Назначение: Быстрое построение графика для данных в объекте Series.
# Типы графиков: kind='line', 'bar', 'hist', 'box', 'pie', 'scatter', и др.
# Особенности: Автоматически использует индекс как ось X, поддерживает настройки Matplotlib через **kwargs.

import pandas as pd
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.plot(kind='line')  # или .plot.line()
plt.show()
```

```python
# DataFrame.plot()  
# Назначение: Построение графиков для данных в DataFrame.
# Особенности: Поддерживает множественные столбцы для построения графиков одновременно.

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df.plot(x='A', y='B', kind='scatter')
plt.show()
```
